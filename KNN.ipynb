# K-nearest neighbors

The K-nears neighbors or the KNN looks at the variable closes to the analyzed variable. In this case the variable: S1 goed which means Symptom 1 correct.

```python
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel("Predictions.xlsx")
df.head(30)
```

```python
dummies = pd.get_dummies(df['Studierichting'])
df = pd.concat([df, dummies], axis=1) #the axis=1 means: add it to the columns (axis=0 is rows)
df.head()
```

```python
X = df[['S2 goed', 'S3 goed', 'S4 goed', 'Aarde en milieu', 'Economie en bedrijf', 'Exact en informatica', 
        'Gedrag en maatschappij', 'Gezondheid', 'Interdisciplinair', 'Kunst en cultuur', 'Onderwijs en opvoeding', 
        'Recht en bestuur', 'Taal en communicatie', 'Techniek']] 

y = df['S1 goed'] 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) 

X_train.head() 
```

```python
from sklearn.neighbors import KNeighborsClassifier 

knn = KNeighborsClassifier(n_neighbors=3)
knn = knn.fit(X_train, y_train) 
```

```python
knn.score(X_test, y_test) 
```

```python
from sklearn.metrics import confusion_matrix
y_test_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_test_pred) 
cm
```
